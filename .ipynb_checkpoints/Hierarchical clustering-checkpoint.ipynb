```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances
```

```python
df=pd.read_csv("gene.csv",header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
from scipy.spatial.distance import pdist, squareform
import math
d=['euclidean','correlation','manhattan']
l=['single','complete','average']
distances = pdist(df)
# Create a square distance matrix
distance_matrix = squareform(distances)
avd={}
dicor1={}
gh=-1
gj=-1
# Calculate the average distance within each cluster
fig, axes = plt.subplots(len(d),len(l), figsize=(15, 15))
for i in d:
    gh=gh+1
    gj=-1
    for j in l:
        gj=gj+1
        xl=[]
        yl=[]
        for n in range(1,10):
            clustering=AgglomerativeClustering(n_clusters=n,linkage=j, metric=i)
            labels=clustering.fit_predict(df)
            dicor1[(i,j)]=labels
            average_distances = []
            unique_labels = np.unique(dicor1[(i,j)])
            for label in unique_labels:
                indices = np.where(dicor1[(i,j)] == label)[0]
                cluster_distances = distance_matrix[np.ix_(indices, indices)]
                average_distance = cluster_distances.mean()
                average_distances.append(average_distance)
            average=np.array(average_distances).mean()
#             print(len(average_distances))
            xl.append(n)
            yl.append(average)
# #         plt.plot(xl,yl)
#         fig,ax=plt.subplots()
#         ax.plot(xl,yl,linestyle='-')
        ax=axes[gh,gj]
#         ax=fig.add_subplot(111,projection='3d')
        ax.plot(xl,yl,linestyle='-')
        ax.set_title(str(i)+" "+str(j))
```

```python
d=['euclidean','correlation','manhattan']
l=['single','complete','average']
dicor={}
for dis_measure in d:
    for linkage in l:
        clustering=AgglomerativeClustering(n_clusters=2,linkage=linkage, metric=dis_measure)
        labels=clustering.fit_predict(df)
        dicor[(dis_measure,linkage)]=labels
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
pc=pca.fit_transform(df)
ev=pca.explained_variance_ratio_
cev=np.cumsum(ev)
print(cev)
```

```python
d=['euclidean','correlation','manhattan']
l=['single','complete','average']
dic={}
for dis_measure in d:
    for linkage in l:
        clustering=AgglomerativeClustering(n_clusters=2,linkage=linkage, metric=dis_measure)
        labels=clustering.fit_predict(pc)
        dic[(dis_measure,linkage)]=labels
```

```python
x=[]
y=[]
z=[]
for i in range(len(pc)):
    x.append(pc[i][0])
    y.append(pc[i][1])
    z.append(pc[i][2])
```

```python
dfdic={}
# dfdic['coordinates']=pc
for i in d:
    for j in l:
        dfdic[(i,j)]=dic[(i,j)]
```

```python
color=['r','b']
gh=-1
gj=-1
def cfunc(i):
    return color[i]
fig, axes = plt.subplots(len(d),len(l), figsize=(15, 15), subplot_kw={'projection': '3d'})
for i in d:
    gh=gh+1
    gj=-1
    for j in l:
        gj=gj+1
        fig=plt.figure()
        ax=axes[gh,gj]
#         ax=fig.add_subplot(111,projection='3d')
        ax.scatter(x,y,z,c=list(map(cfunc,df1[(i,j)])),marker="o")
        ax.set_title(str(i)+" "+str(j))
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, completeness_score, homogeneity_score, v_measure_score
dicscore={}
lsilhouette=[]
ldavies_bouldin=[]
lcalinski_harabasz=[]
lt=[]
for i in d:
    for j in l:
        silhouette = silhouette_score(df, dicor[(i,j)])
        davies_bouldin = davies_bouldin_score(df, dicor[(i,j)])
        calinski_harabasz = calinski_harabasz_score(df, dicor[(i,j)])
        lsilhouette.append(silhouette)
        ldavies_bouldin.append(davies_bouldin)
        lcalinski_harabasz.append(calinski_harabasz)
        lt.append([i,j])
#         print("silhoutette score",silhouette,"davies_bouldin",davies_bouldin,"calinski_harabasz",calinski_harabasz,f"for {i,j}")

dicscore["silhoutte score"]=lsilhouette
dicscore["davies_bouldin"]=ldavies_bouldin
dicscore["calinski_harabasz"]=lcalinski_harabasz
dicscore["dis_measure and linkage"]=lt
dscore=pd.DataFrame(dicscore)
print(dscore)
```
